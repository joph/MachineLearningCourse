```python
import folium
print(folium.__version__)

# Define the URL format used for Earth Engine generated map tiles.
EE_TILES = 'https://earthengine.googleapis.com/map/{mapid}/{{z}}/{{x}}/{{y}}?token={token}'
```

```python
bands = ['B2', 'B3', 'B4', 'B5', 'B6', 'B7']
#bands = ['B2', 'B3', 'B4']
# Use Landsat 8 surface reflectance data.
l8sr = ee.ImageCollection("LANDSAT/LE07/C01/T1_SR")

# Cloud masking function.
def maskL8sr(image):
  cloudShadowBitMask = ee.Number(2).pow(3).int()
  cloudsBitMask = ee.Number(2).pow(5).int()
  qa = image.select('pixel_qa')
  mask = qa.bitwiseAnd(cloudShadowBitMask).eq(0).And(
    qa.bitwiseAnd(cloudsBitMask).eq(0))
  return image.updateMask(mask).select(bands).divide(10000)

# The image input data is a 2018 cloud-masked median composite.
reducer = ee.Reducer.median()

reducer_str = "median"

years = range(2000, 2019)

images = []

for year in years:
    image = l8sr.filterDate(str(year) + '-01-01', str(year) + '-12-31').map(maskL8sr).reduce(reducer)
    images.append(image)
    
```

```python
import ee.mapclient

vals = []

for i in range(0,19):
    image_final = images[i]
    
    
    x = 16.945703049327297
    
    y = 47.911432140538195
    
    p = ee.Geometry.Point(x, y)

    # Extract the data
    data = image_final.select("B2_median").reduceRegion(ee.Reducer.first(), p, 30).get("B2_median")

    # Convert to Number for further use
    dataN = ee.Number(data)
    
    d = dataN.getInfo()
    
    vals.append(d)
    
    # Show data
    print(d)
 
    mapid = image_final.getMapId({'bands': ['B2_' + reducer_str], 'min': 0, 'max': 0.3})
    map = folium.Map(location=[y, x], zoom_start=25)
    
    folium.TileLayer(
        tiles=EE_TILES.format(**mapid),
        attr='Google Earth Engine',
        overlay=True,
        name='median composite',
      ).add_to(map)
    
    folium.vector_layers.Circle((y, x), 10).add_to(map)
    
    display(map)
    
    
    
    #mapid = image_final.getMapId({'bands': ['B2_' + reducer_str], 'min': 0, 'max': 0.3})
    
#, "B3_" + reducer_str, "B4_" + reducer_str
    #mapid = m1.getMapId()

    #map = folium.Map(location=[48.196695, 16.679343], zoom_start=25)
    #folium.TileLayer(
    #    tiles=EE_TILES.format(**mapid),
    #    attr='Google Earth Engine',
    #    overlay=True,
    #    name='median composite',
    #  ).add_to(map)
    #map.add_child(folium.LayerControl())
    #display(map)
```

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

def plotturbine(year, images):
    offset = 0.001
    
    y = 16.945403049327297
    x = 47.911632140538195
    
    rectangle = ee.Geometry.Polygon([
                                    [x - offset, y - offset],
                                    [x - offset, y + offset],
                                    [x + offset, y + offset],
                                    [x + offset, y - offset]
                                    ])
    
    #rectangle = ee.Geometry.Rectangle([16.679343 - offset, 48.196695 - offset, 16.679343 + offset, 48.196695 + offset])
    colReducer = ee.Reducer.toList()
    col = images[year].reduceRegion(reducer = colReducer, geometry = rectangle, scale = 30)
    
    data = np.array((ee.Array(col.get("B2_median")).getInfo())).reshape(7, 7)
    
    #print(data)
    
    import matplotlib.image as mpimg
    plt.imshow(data)
    plt.figure()
    return(data)

images_np = []

for i in range(15, 19):
    img = plotturbine(i, images)
    images_np.append(img)
    

    
    

```
